# Changing the point style

```python
markers = [".", ",", "o", "v", "^", "<", ">", "1", "2", "3", "4", "8", "s", "p", "P", "*", "h", "H", "+", "x", "X", "D", "d", "|", "_"]

for i, m in enumerate(markers):
    plt.scatter(np.random.random(1), i, marker=m, s=100, color='black', label=m)

plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', prop={'size': 12})
plt.title('Random scatter plot with random markers')
```

# Adding more information

It is quite common that the two dimensions of the scatter plot will not be sufficient to display all the information you want. For example, if you wanted to compare the average height, weight and age of a set of patents. You could use the `x` axis to denote their height, the `y` axis to denote their weight and the size of the point to denote their age. I would not advise using the size of the point to denote their weight---someone might get upset.

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
s = np.random.rand(1000) * 500

plt.scatter(x, y, s, color='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random scatter plot with size as a variable')
```

## Dealing with overlapping points

If the points overlap you have two, good, options. First, you can use the `alpha` parameter to add transparency. However, this will not work for some file format (e.g. `pdf`).

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
s = np.random.rand(1000) * 500

plt.scatter(x, y, s, color='black', alpha=0.2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random scatter plot with size as a variable')
```

If, for whatever reason, you dont want to use the `alpha` parameter, you can just make all the points hollow. This can be done by setting the `facecolors` parameter to `'none'` (not `None`!).

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
s = np.random.rand(1000) * 500

plt.scatter(x, y, s, color='black', facecolors='none', edgecolors='black')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random scatter plot with size as a variable')
```

## Adding a colorbar

Although using the size of the point can be a quick way to show extra information about each point, it is not trivial to compare the sizes of two points. If we want to report this information more accurately, we can change from using the size to using a colormap. In the figure below I use the `z` variable to control the size and color of the points.

```python
x = np.random.rand(1000)
y = np.random.rand(1000)
z = np.random.rand(1000) * 500

# convert the z variable into a list of colors
# cm_z = np.linspace(0.0, 1.0, len(z))
normalized_z = z / np.max(z)
colors = [cm.viridis(x) for x in normalized_z]

plt.scatter(x, y, z, color=colors)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random scatter plot with size as a variable')
plt.colorbar(label='z')
```
